```python
#dilate扩大函数
import cv2
import numpy as np
#读取图像
src=cv2.imread( 'pengzhang.png' ,cv2.IMREAD_UNCHANGED)
#设置卷积核
kernel=np.ones( (5,5) ,np.uint8)
#图像膨胀处理
erosion=cv2.dilate(src,kernel)
#显示图像
cv2.imshow("src",src)
```

```python
cv2.imshow( "result",erosion)
#等待显示
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('pengzhang.png',0)

# 创建掩膜
mask = np.zeros(img.shape[:2], np.uint8)#生成全黑图像
mask[50:300, 140:400] = 255#部分取白
masked_img = cv2.bitwise_and(img,img,mask = mask)
# 分别计算有无掩膜情况下的直方图
# Check third argument for mask
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])#不使用掩膜，只对第一通道
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])#使用掩膜，只对第一通道

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```
