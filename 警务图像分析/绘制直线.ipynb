```python
import cv2
import numpy as np

# 创建一个空白图像
img = np.zeros((256, 256, 3), dtype=np.uint8)

i = 0
while i < 255:
    # 绘制直线
    cv2.line(img, (0, i), (255, 255 - i), (55, 255, 155), 5)
    i = i + 1

# 显示图像
cv2.imshow('lines', img)

# 等待显示
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#读取图像
img=cv2.imread('test3.bmp')
src=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#图像翻转
img1=cv2.flip(src,0)#参数=0以X轴为对称轴翻转
img2=cv2.flip(src,1)#参数>0以Y轴为对称轴翻转
img3=cv2.flip(src,-1)#参数<0以X轴和Y轴为对称轴翻转
#显示图像
titles=['source','Image1','Image2','Image3']
images=[src,img1,img2,img3]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
