```python
import cv2
import numpy as np

# 读取图像 xianyiren.jpg
src = cv2.imread("xianyiren.jpg")

# 获取图像的高度和宽度
height, width = src.shape[:2]

# 定义旋转矩阵，实现180度旋转
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 180, 1)

# 进行旋转变换
rotated_image = cv2.warpAffine(src, rotation_matrix, (width, height))

# 保存图像为 xianyiren_rotated.jpg
cv2.imwrite("xianyiren_rotated.jpg", rotated_image)
```

```python
import cv2

# 读取图像 xianyiren.jpg
src = cv2.imread("xianyiren1.jpg")

# 获取原始图像的高、宽以及通道数
rows, cols, channel = src.shape

# 定义旋转矩阵
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), -40, 1)

# 对输入图像按旋转矩阵进行旋转
rotated = cv2.warpAffine(src, M, (cols, rows))

# 将旋转后的图像的宽和高都缩小为原来的一半
resized = cv2.resize(rotated, (cols // 2, rows // 2))

# 将图像保存为 xianyiren2.jpg
cv2.imwrite("1667890875060004445.jpg", resized)
```
